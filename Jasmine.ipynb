```python
# Function to extract text from specific pages
def extract_specific_pages(pdf_path, start_page, end_page):
    with pdfplumber.open(pdf_path) as pdf:
        extracted_text = ""
        for page_num in range(start_page, end_page):
            page = pdf.pages[page_num]
            extracted_text += page.extract_text() + "\n"
        return extracted_text
```

```python
def find_relevant_section(query, qa_mapping):
    query = query.lower().strip()
    
    # Direct match
    if query in qa_mapping:
        return qa_mapping[query]
    
    # Partial match
    for key_phrase in qa_mapping:
        if key_phrase in query or query in key_phrase:
            return qa_mapping[key_phrase]
    
    return "No matching section found"
```

```python
# Example usage
test_queries = [
    "What is the authority of the president?",
    "What does the president do?",
    "Who are principal secretaries?",
    "What is the cabinet?",
    "How does the president make decisions?"
]

for query in test_queries:
    section = find_relevant_section(query, qa_mapping_chapter_9)
    print(f"\nQuery: {query}")
    print(f"Relevant section: {section}")
```

```python
def get_section_content(query, chapter_sections, qa_mapping):
    relevant_section = find_relevant_section(query, qa_mapping)
    if relevant_section in chapter_sections:
        return f"Section: {relevant_section}\n\n{chapter_sections[relevant_section]}"
    return "Sorry, I couldn't find relevant information for your query."

# Example usage
query = "What is the authority of the president?"
response = get_section_content(query, chapter_9_sections, qa_mapping_chapter_9)
print(response)
```
