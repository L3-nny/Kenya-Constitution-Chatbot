### creating a function that will help to iterate text using the page number 

```python
# Path to your PDF file
pdf_path = "C:/Users/PC/Documents/FLATIRON/PHASE5/Kenya-Constitution-Chatbot/Kenya Constitution.pdf"

# Function to extract text from specific pages
def extract_specific_pages(cleaned_constitution, start_page, end_page):
    with pdfplumber.open(cleaned_constitution) as pdf:
        extracted_text = ""
        for page_num in range(start_page, end_page):
            page = pdf.pages[page_num]
            extracted_text += page.extract_text() + "\n"
        return extracted_text
```

```python
# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Preprocess the user query using spaCy
def preprocess_query(query):
    # Parse the query with spaCy
    doc = nlp(query)
    # Normalize the query: lowercase, lemmatize, and remove stopwords
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)

# Example usage of the preprocessing function
user_query = "What is Bill of rights?"
processed_query = preprocess_query(user_query)
print(processed_query)
```

```python
# extracting the whole of chapter 4 which is located in page 18 to 41
# chapter_4 = extract_specific_pages(pdf_path, 18,41 )
# chapter_4
#Function to clean extracted text
def clean_extracted_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s\n]', ' ', text)  # Remove non-alphabetic characters
    return text

# Extract and clean chapter 4
chapter_4 = extract_specific_pages(pdf_path, 18, 41)
cleaned_chapter_4 = clean_extracted_text(chapter_4)
print(cleaned_chapter_4)
```

```python
def split_chapter(chapter_text):
    # Split at key headings and strip extra whitespace
    sections = {
        "rights and fundamental freedoms": [],
        "application": [],
        "implementation of rights and fundamental freedoms": [],  
        "enforcement of bill of rights": [],
        "authority of courts to uphold and enforce the Bill of Rights": [],
        "limitation of rights or fundamental freedoms": [],
        "fundamental Rights and freedoms that may not be limited": [],
        "right to life": [],
        "equality and freedom from discrimination": [],
        "human dignity": [],
        "freedom and security of the person": [],
        "slavery, servitude and forced labour": [],
        "privacy": [],
        "freedom of conscience, religion, belief and opinion": [],
        "freedom of expression": [],
        "freedom of the media": [],
        "access to information": [],
        "freedom of association": [],
        "assembly, demonstration, picketing and petition": [],
        "political rights": [],
        "freedom of movement and residence": [],
        "protection of right to property": [],
        "labour relations": [],
        "environment": [], 
    }

    # Split by new lines to process line by line
    lines = chapter_text.splitlines()

    current_section = None

    for line in lines:
        stripped_line = line.strip()

        if stripped_line.startswith("rights and fundamental freedoms"):
            current_section = "rights and fundamental freedoms"
        elif stripped_line.startswith("application of bill of rights"):
            current_section = "application"
        elif stripped_line.startswith("implementation of rights and fundamental freedoms"):
            current_section = "implementation of rights and fundamental freedoms"
        elif stripped_line.startswith("enforcement of bill of rights"):
            current_section = "enforcement of bill of rights"
        elif stripped_line.startswith("authority of courts to uphold and enforce the Bill of Rights"):
            current_section = "authority of courts to uphold and enforce the Bill of Rights"
        elif stripped_line.startswith("limitation of rights or fundamental freedoms"):
            current_section = "limitation of rights or fundamental freedoms"
        elif stripped_line.startswith("right to life"):
            current_section = "right to life"
        elif stripped_line.startswith("equality and freedom from discrimination"):
            current_section = "equality and freedom from discrimination"
        elif stripped_line.startswith("human dignity"):
            current_section = "human dignity"
        elif stripped_line.startswith("freedom and security of the person"):
            current_section = "freedom and security of the person"    

        # Append line to the current section if it's set
        if current_section:
            sections[current_section].append(stripped_line)

    # Join each section into a single string
    for key in sections:
        sections[key] = "\n".join(sections[key])

    return sections

# Split the chapter into sections
chapter_4_sections = split_chapter(chapter_4_trimmed)

print("rights and fundamental freedoms Section:\n", chapter_4_sections['rights and fundamental freedoms'])
print("\napplication Section:\n", chapter_4_sections['application'])
print("\nimplementation Section:\n", chapter_4_sections['implementation of rights and fundamental freedoms'])

```

```python
import spacy

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Preprocess the user query using spaCy
def preprocess_query(query):
    # Parse the query with spaCy
    doc = nlp(query)
    # Normalize the query: lowercase, lemmatize, and remove stopwords
    tokens = [token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)

# Define the sections variable with Chapter 1
sections = [chapter_4_trimmed]  # Assuming chapter_1 contains the text extracted from the PDF

# Define the QA mapping based on key phrases and corresponding sections
qa_mapping = {
    "rights": "rights and fundamental freedoms",
    "application": "application of bill of rights",
    "implementation": "implementation of rights and fundamental freedoms"
    # Add more mappings as needed
}

# Update the Q&A system to use preprocessed queries
def answer_question_nlp(query, sections, qa_mapping):
    # Preprocess the user query
    processed_query = preprocess_query(query)
    
    # Search for a key in qa_mapping that matches the preprocessed query
    for key in qa_mapping:
        if key in processed_query:
            # Find the relevant section in the text
            for section in sections:
                if qa_mapping[key].lower() in section.lower():
                    return section
    return "Sorry, I couldn't find an answer to your question."

# Example usage
user_query = "What is the Implementation of rights and fundamental freedoms?"
answer = answer_question_nlp(user_query, sections, qa_mapping)
print(answer)
```
